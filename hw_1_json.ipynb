## Вопрос 1
Каждый элемента файла - 1 твит, значит мы ищем длину файла. Именно значение длины файла и есть общее количество твитов

```python
import json
from collections import Counter
from pprint import pprint

twitter = []
for line in open('hw_3_twitter.json', 'r', encoding='utf-8'):
    twitter.append(json.loads(line))
```

**Ответ**

```python
print(len(twitter))
```

## Вопрос 2
Удаленные твиты имеют ключ 'delete'. Пройдемся по всем твитам. Добавим в пустой список все удаленные твиты (поставим условие, что твит имеет тег 'delete'), а затем найдем длину этого списка. Это будет количество удаленных твитов. После чего рассчитаем процент удаленных твитов.

```python
deleted = []
for twit in twitter:
    if 'delete' in twit:
        deleted.append(twit)
```

**Ответ**

```python
answer = (len(deleted) * 100) / len(twitter)
print(answer)
```

## Вопрос 3
Для решения задачи нужно найти все языки, на которых написаны твиты. Пройдемся по всем твитам. Затем проверим твит на наличие языка (язык имеет ключ 'lang'), если значение есть, то мы добавляем язык в общий список. После чего подключим Counter к списку всех найденных языков и выведем топ-15.

```python
languages = []
for twit in twitter:
    lang = twit.get('lang')
    if lang is not None:
        languages.append(lang)
languages = Counter(languages)
```

**Ответ**

```python
languages.most_common(15)
```

## Вопрос 4
Будем искать пользователей по их id, так как он у каждого пользователя уникальный. Проверим твит на наличие информации о пользователе (ключ 'user' отвечает за это). И если информация есть, то мы добавим в наш список всех пользователей id каждого из них. Подлючим Counter и добавим те id, которые встречаются больше 1 раза, в новый список повторяющихся пользователей. После чего найдем длину списка - количество пользователей, написавших больше одного твита.

```python
users = []
for twit in twitter:
    user = twit.get('user')
    if user is not None:
        users.append(user.get('id'))
repetitive_names = []
names = Counter(users)
for user in names:
    if names[user] > 1:
        repetitive_names.append(user)
```

**Ответ**

```python
len(repetitive_names)
```

## Вопрос 5
Чтобы добраться к хэштэгам, нужно пройтись по твитам, найти ключ 'entities', и если он есть, то нужно проверить на наличие в его значении ключа 'hashtags'. Добавим текст хэштэгов, где он есть, в список. Подключим Counter и выведем топ-20 хэштэгов.

```python
hashtags = []
for twit in twitter:
    entities = twit.get('entities')
    if entities is not None:
        hashtag = entities.get('hashtags')
        if hashtag is not None:
            for i in hashtag:
                text = i['text']
                hashtags.append(text)
hashtags = Counter(hashtags)
```

**Ответ**

```python
hashtags.most_common(20)
```

## Вопрос 6
Для начала выберем все твиты, которые написаны только на английском языке. Затем исключим с помощью условий ретвиты, оставив только оригинальные твиты. Приведем твиты к нижнему регистру. Избавимся от пунктуации и добавим отредактированые твиты в список.

```python
import string

texts = []
for twit in twitter:
    if twit.get('lang') == 'en':
        if twit.get('retweeted_status') is None:
            if twit.get('is_quote_status') is False:
                text = twit.get('text').lower()
                text = text.translate(text.maketrans
                                      ('', '', string.punctuation))
                texts.append(text)
```

Создадим список для отдальных слов и добавим в него слова из твитов, поделив каждый твит на слова с помощью split по пробелу. Затем применим Counter к списку слов и сделаем частотный словарь.

```python
words = []
for twit in texts:
    words.extend(twit.split(' '))
```

**Ответ**

```python
Counter(words).most_common()
```

## Вопрос 7
Информация о подписчиках содержится в ключе 'user', поэтому мы проверяем наличие этой информации. Если она есть, то мы создаем кортеж с именем пользователя и количеством подписчиков. Добавляем этот кортеж в список. И делаем сортировку, выводя топ-10 пользователей с наибольшим числом подписчиков.

```python
followers = []
for twit in twitter:
    user = twit.get('user')
    if user is not None:
        all = (user.get('name'), user.get('followers_count'))
        followers.append(all)
```

**Ответ**

```python
from operator import itemgetter

sorted(followers, key=itemgetter(1), reverse=True)[:10]
```

## Вопрос 8*
Проходясь по твитам, мы находим те, которые имеют информации об источнике твита (по ключу 'source'). Если она есть, то с помощью регулярного выражения мы находим название приложений, заключенных в >< и добавляем их в список. Подключаем к списку Counter и выводим топ-10 источников, используеымых для поста твитов.

```python
import re

applications = []
for twit in twitter:
    source = twit.get('source')
    if source is not None:
        apl = re.findall(r'>(.+[^<])<', source)
        applications.append(apl[0])
```

**Ответ**

```python
Counter(applications).most_common(10)
```
